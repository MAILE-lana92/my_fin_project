```python
import os, openpyxl, sys
from os import listdir
import pandas as pd, numpy as np
file_path = r"D:\working\data_projects\converted\three_years_history"
save_path = r"D:\my_fin_project\seeds\base_transaction.csv"
header = ['No','Date','Code','Desc','Credit','Debit','Balance']
```

```python
file_list = os.listdir(file_path)
```

```python
def concat_df(file_path):
    df = pd.DataFrame(columns= header)
    for file in file_list:
        sheets = pd.ExcelFile(os.path.join(file_path, file)).sheet_names
        print(os.path.join(file_path, file))
        print(sheets)
        for sheet in sheets:
            df_batch = pd.read_excel(os.path.join(file_path, file), sheet_name= sheet, skiprows= 1, header= None)
            df_batch.columns = header
            df = pd.concat([df, df_batch], ignore_index= True)
            print(df_batch.shape)
            print(df.shape)
    return df
        
```

```python
def clean_df(df_concat):
    # dedup dataframe
    df_dedup = df_concat.drop_duplicates(subset= header, keep= 'first') 
    # remove records with invalid value
    df = df_dedup[df_dedup['Balance'].notnull()]
    df = df[df['Date'] != 'Ngày/Date'] 
    df = df[df_dedup['Date'] != 'Ngày|Date'] 
    # convert not NA value to numeric
    df['No'] = df['No'].apply(lambda x: pd.to_numeric(x, errors='coerce'))
    # with value NaN in row n, replace by value in row n - 1
    df['No'] = df['No'].fillna(method='ffill') + df['No'].isnull().astype(int)
    # convert Date column to correct date format
    df['Date'] = df['Date'].astype('datetime64[ns]').dt.strftime('%Y-%m-%d')
    # remove invalue characters in convert Credit, Debit. Balance columns into int
    df['Credit'] = df['Credit'].astype(str).str.replace(r'[^0-9]', '', regex= True).astype(int)
    df['Debit'] = df['Debit'].astype(str).str.replace(r'[^0-9]', '', regex= True).astype(int)
    df['Balance'] = df['Balance'].astype(str).str.replace(r'[^0-9]', '', regex= True).astype(int)
    # replace Enter pattern with ' '
    df['Desc'] = df['Desc'].replace('\n', ' ', regex= True)
    return df
```

```python


def get_sheet1_data(file_path, file_name):
    try:
        header = ['No','Date','Code','Desc','Credit','Debit','Balance']
        df = pd.read_excel(f'{file_path}\{file_name}.xlsx', sheet_name='Sheet1', engine='openpyxl',skiprows=1,header=None)
        df.columns = header
        df.iloc[:,3]= df.iloc[:,3].astype(str).str.replace('\n', ' ', regex=False) 
        df.iloc[:,3] = df.iloc[:,3].astype(str).str.replace('\r', ' ', regex=False)  
        df.iloc[:,4] = df.iloc[:,4].astype(str).str.replace('\n', ' ', regex=False) 
        df.iloc[:,5] = df.iloc[:,5].astype(str).str.replace('\n', ' ', regex=False) 
        # df.iloc[:,6]  = df.iloc[:,6]. .astype(str).str.replace('\n', ' ', regex=False) 
        df.iloc[:,4] = df.iloc[:,4].astype(str).str.replace(',', '', regex=False) 
        df.iloc[:,5] = df.iloc[:,5].astype(str).str.replace(',', '', regex=False) 
        # df.iloc[:,6]  = df.iloc[:,6]. .astype(str).str.replace(',', '', regex=False) 
        return df
    except Exception as e:
        print(f'Error: {e}')
        
# def get_first_row(file_path, file_name):
#     try:
#         df = pd.read_excel(f'{file_path}\{file_name}.xlsx', sheet_name='Sheet1', engine='openpyxl',header=0)
#         first_transdate =.astype(str) str(df['STT/No'][0].astype.astype(str)(str))
#         return first_transdate
#     except Exception as e:
#         print(f'Error: {e}')

def get_sheet_list(file_path, file_name):
    xls = pd.ExcelFile(f'{file_path}\{file_name}.xlsx')
    sheet_list= xls.sheet_names[1:]
    return sheet_list

def save_ingest_data(file_path, file_name):
    try:
        get_sheet1_data(file_path, file_name).to_csv(os.path.join(save_path,f'ingest_data.csv'), mode= 'a', index=False, header=False)
        for sheet in get_sheet_list(file_path, file_name):
            df = pd.read_excel(f'{file_path}\{file_name}.xlsx', sheet_name=sheet, engine='openpyxl',skiprows=1, header=None)
            df.iloc[:,3]= df.iloc[:,3].astype(str).str.replace('\n', ' ', regex=False) 
            df.iloc[:,3] = df.iloc[:,3].astype(str).str.replace('\r', ' ', regex=False)  
            df.iloc[:,4] = df.iloc[:,4].astype(str).str.replace('\n', ' ', regex=False) 
            df.iloc[:,5] = df.iloc[:,5].astype(str).str.replace('\n', ' ', regex=False) 
            # df.iloc[:,6] = df.iloc[:,6].astype(str).str.replace('\n', ' ', regex=False) 
            df.iloc[:,4] = df.iloc[:,4].astype(str).str.replace(',', '', regex=False) 
            df.iloc[:,5]  = df.iloc[:,5].astype(str).str.replace(',', '', regex=False) 
            # df.iloc[:,6]  = df.iloc[:,6]. .astype(str).str.replace(',', '', regex=False) 
            df = df.drop_duplicates()
            df = df.loc[df.iloc[:,1] != 'Ngày/Date']
            df.to_csv(os.path.join(save_path,f'ingest_data.csv'),mode='a', index=False,header=False)
            print(f'{sheet} has been ingested to {save_path}')
    except Exception as e:
        print(f'Error: {e}')

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python ingest_excel_data.py <file_name>")
        sys.exit(1)
    
    file_name = sys.argv[1]

save_ingest_data(file_path, file_name)  
```
